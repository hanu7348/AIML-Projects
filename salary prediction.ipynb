```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance for each categorical variable
le = LabelEncoder()

# Apply the encoding and reassign to the DataFrame
salary['Education Level'] = le.fit_transform(salary['Education Level'])
salary['Gender'] = le.fit_transform(salary['Gender']) 
salary['Job Title'] = le.fit_transform(salary['Job Title']) 

# Prepare the data (after encoding)
X = salary[['Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience']] 
y = salary['Salary']

#create an empty model
model = LinearRegression()
model.fit(X, y)
```

```python
#user input
age=float(input("Enter age: "))
gen=int(input("Enter0 for male 1 for female: "))
el=int(input("Enter education level 0-Bachelors,1-Master,2-PhD: "))
exp=float(input("Enter years of experience: "))
out=model.predict([[age,gen,el,exp]])
print("Your Predicted salary is:",out[0])
```
